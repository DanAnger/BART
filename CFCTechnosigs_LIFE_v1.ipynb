#Initial conditions and observation options

```
#@title 

"""
The class Options contains the initial conditions
"""
class Options:
    def __init__(self, image_size=512):
        '''
        options to choose
        '''
        #paramters for wl sweep
        self.t_tot = 3600 * 1    # seconds
        self.wl_optimal = 20e-6 # in m
        self.lz_model = "darwinsim"
        #self.lz_model = "glasse"


        '''
        options derived from inputs.
        '''       
        self.image_size = image_size  
        
        # coordinate maps for faster calculations
        self.x_map = np.tile(np.array(range(0, self.image_size)),
                             (self.image_size,1))
        self.y_map = self.x_map.T
        self.r_square_map = ((self.x_map - (self.image_size-1) / 2)**2
                            +(self.y_map - (self.image_size-1) / 2)**2)
        self.r_map = np.sqrt(self.r_square_map)
        


options = Options(image_size=256)

"""
main
"""



inst = Instrument(options, D=2,
                  quantum_eff=0.7,
                  throughput=0.05,
                  wl_min=4.,
                  wl_max=18.5,
                  R=50,
                  bl_min=10,
                  bl_max=100)

```
